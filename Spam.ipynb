```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
```

```
spam_keyword_url = 'https://raw.githubusercontent.com/pranjal3163/datascience/master/dataset_44_spambase.csv'
df = pd.read_csv(spam_keyword_url)
print(df)
```

```
target = df.pop("class")
target
```

```
seed = 168
rf = RandomForestClassifier(random_state= seed)
rf.fit(df,target)
```

```
preds = rf.predict(df)
print(preds)
```

```
accuracy_score(target,preds)
```
